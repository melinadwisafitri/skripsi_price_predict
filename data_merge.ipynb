## Data Load

```python
url_2017 = 'https://github.com/melinadwisafitri/skripsi_price_predict/blob/main/data_pangan/2017.xlsx?raw=true'
url_2018 = 'https://github.com/melinadwisafitri/skripsi_price_predict/blob/main/data_pangan/2018.xlsx?raw=true'
url_2019 = 'https://github.com/melinadwisafitri/skripsi_price_predict/blob/main/data_pangan/2019.xlsx?raw=true'
url_2020 = 'https://github.com/melinadwisafitri/skripsi_price_predict/blob/main/data_pangan/2020.xlsx?raw=true'
url_2021 = 'https://github.com/melinadwisafitri/skripsi_price_predict/blob/main/data_pangan/2021.xlsx?raw=true'
url_2022 = 'https://github.com/melinadwisafitri/skripsi_price_predict/blob/main/data_pangan/2022.xlsx?raw=true'

urllib.request.urlretrieve(url_2017, 'price_2017.xlsx')
urllib.request.urlretrieve(url_2018, 'price_2018.xlsx')
urllib.request.urlretrieve(url_2019, 'price_2019.xlsx')
urllib.request.urlretrieve(url_2020, 'price_2020.xlsx')
urllib.request.urlretrieve(url_2021, 'price_2021.xlsx')
urllib.request.urlretrieve(url_2022, 'price_2022.xlsx')
```

### Merge all Data

```python
merge_data = pd.concat([data_2017, data_2018, data_2019, data_2020, data_2021, data_2022], axis=1, join="outer")
merge_data = merge_data.drop(columns='No.')
```

#### Transform merge Data

```python
merge_data_transpose = merge_data.T.reset_index()
```

```python
merge_data_transpose.head()
```

```python
merge_data_transpose.columns = merge_data_transpose.loc[0]
merger = merge_data_transpose.drop(0)
merger
```

```python
merger = merger.rename(columns = {'Komoditas(Rp)': 'Date'})
merger.head()
```

### Export Data

```python
name_file_export = 'data_harga_bahan_pangan_indonesia.csv'
data_final =  merger.to_csv(name_file_export, index=False)
files.download(name_file_export)
```
